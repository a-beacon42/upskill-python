# Module 1 — Build From Scratch: Grade Calculator

**Write a function `grade_summary(raw_scores: list[str]) -> dict` that:**

1. Accepts a list of scores as **strings** (e.g. `["85", "92", "foo", "73"]`)
2. Skips any value that can't be converted to a number
3. Clamps valid scores to the range 0–100
4. Maps each valid score to a letter grade:
   - 90–100 → `"A"`, 80–89 → `"B"`, 70–79 → `"C"`, 60–69 → `"D"`, below 60 → `"F"`
5. Returns: `{"grades": ["B", "A", "C"], "average": 83.3, "invalid_count": 1}`

**Rules:**

- Extract the mapping into a helper: `score_to_grade(score: int) -> str`
- Round the average to 1 decimal place
- Empty list → `{"grades": [], "average": 0.0, "invalid_count": 0}`

**Optional stretch:** Add a `print_report` helper that prints a formatted summary.


```python
def clamp(value: float, low: float, high: float) -> float:
    if value < low:
        return low
    if value > high:
        return high
    return value


def validate_raw_scores(raw_scores: list[str]) -> tuple:
    valid_grades: list = []
    invalid_count: int = 0
    for score in raw_scores:
        try:
            valid_grade = clamp(int(score), 0, 100)
            valid_grades.append(valid_grade)
        except ValueError:
            invalid_count += 1
    return valid_grades, invalid_count


def map_to_letter_grade(grade: int) -> str:
    if grade >= 90:
        return "A"
    if grade >= 80:
        return "B"
    if grade >= 70:
        return "C"
    if grade >= 60:
        return "D"
    return "F"


def get_avg(number_scores: list[int]) -> float:
    return (
        round(sum(number_scores) / len(number_scores), 1)
        if len(number_scores) > 0
        else 0.0
    )
```

```python
# Your solution here
def grade_summary(raw_scores: list[str]) -> dict:
    valid_grades, invalid_count = validate_raw_scores(raw_scores)
    letter_grades = [map_to_letter_grade(grade) for grade in valid_grades]
    average = get_avg(valid_grades)

    return {"grades": letter_grades, "average": average, "invalid_count": invalid_count}
```

```python
# Test cases — run these after you've written your solution
print(grade_summary(["85", "92", "foo", "73"]))
# Expected: {'grades': ['B', 'A', 'C'], 'average': 83.3, 'invalid_count': 1}

print(grade_summary([]))
# Expected: {'grades': [], 'average': 0.0, 'invalid_count': 0}

print(grade_summary(["110", "-5", "abc", "60"]))
# Expected: {'grades': ['A', 'F', 'D'], 'average': 55.0, 'invalid_count': 1}
```

```python
def grade_to_letter(grade: int) -> str:
    if grade >= 90:
        letter_grade = "A"
    elif grade >= 80:
        letter_grade = "B"
    elif grade >= 70:
        letter_grade = "C"
    elif grade >= 60:
        letter_grade = "D"
    else:
        letter_grade = "F"
    return letter_grade


def validate_score(raw_score) -> int | None:
    try:
        converted_score = int(raw_score)
    except ValueError:
        return None
    if converted_score < 0:
        converted_score = 0
    if converted_score > 100:
        converted_score = 100
    return converted_score


def grade_summary(raw_scores: list[str]) -> dict:
    scores = []
    invalid_count = 0
    total = 0
    for score in raw_scores:
        converted_score = validate_score(score)
        if converted_score is not None:
            letter_grade = grade_to_letter(converted_score)
            scores.append(letter_grade)
            total = total + converted_score
        else:
            invalid_count += 1
    if len(scores) == 0:
        average = 0.0
    else:
        average = round(total / len(scores), 1)
    return {"grades": scores, "average": average, "invalid_count": invalid_count}
```
