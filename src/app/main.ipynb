Given a list of orders, write a function that returns a dictionary mapping each product to its total quantity ordered.


```python
from collections import Counter, defaultdict

orders = [
    {"product": "widget", "quantity": 3},
    {"product": "gadget", "quantity": 1},
    {"product": "widget", "quantity": 7},
    {"product": "gadget", "quantity": 4},
    {"product": "doohickey", "quantity": 2},
]


def total_by_product(orders: list[dict]) -> dict[str, int]:
    total_counts_per_product: dict[str, int] = {}
    for order in orders:
        prod = order["product"]
        total_counts_per_product[prod] = (
            total_counts_per_product.get(prod, 0) + order["quantity"]
        )
    return total_counts_per_product


def total_by_product_counter(orders: list[dict]) -> dict[str, int]:
    total_counts_per_product = Counter()
    for order in orders:
        total_counts_per_product[order["product"]] += order["quantity"]
    return total_counts_per_product


def total_by_product_defaultdic(orders: list[dict]) -> dict[str, int]:
    total_counts_per_product = defaultdict(int)
    for order in orders:
        total_counts_per_product[order["product"]] += order["quantity"]
    return total_counts_per_product


# Expected output: {"widget": 10, "gadget": 5, "doohickey": 2}
print(total_by_product(orders))
print(total_by_product_counter(orders))
print(total_by_product_defaultdic(orders))
```
